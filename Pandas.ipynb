```python
import csv
csv_columns=['Name','Age','Address']
dict = [
    {'Name':'Aayush','Age':23,'Address':'Kathmandu'},
    {'Name':'Shyam','Age':33,'Address':'Patan'}
]
with open("dataset4.csv","w") as file:
    writer=csv.DictWriter(file, fieldnames=csv_columns)
    writer.writeheader()
    for data in dict:
        writer.writerow(data)
```
